```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
```

```python
data=pd.read_csv("/content/House Price India.csv")
data.head()
```

```python

plt.figure(figsize=(20, 10))
data.Price.plot()
plt.show()
```

```python
plt.figure(figsize=(16, 8))
data["Number of schools nearby"].plot()
data["Price"].plot()
plt.legend(["Number of schools nearby",'Price'])
```

```python
plt.figure(figsize=(16, 8))
data["lot area"].plot()
data["living area"].plot()
plt.legend(["lot area",'living area'])
```

```python
sns.distplot(data.Price)
plt.show()
```

```python
import matplotlib.pyplot as plt

condition_counts = df.value_counts()


plt.pie(condition_counts, labels=['3', '2', '1'], autopct='%1.1f%%', shadow=True)
plt.title('no of Schools nearby')
plt.show()

```
